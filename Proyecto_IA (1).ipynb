# Acronimos

FAVC: Alimentos de Alto Valor Calórico Consumidos Frecuentemente

FCVC: La frecuencia de consumo de verduras es una variable que muestra con qué frecuencia una persona come verduras

NCP: Número de comidas principales

CAEC: Consumo de alimentos entre comidas

CH20: Consumo de agua (En litros)

CALC: Consumo de alcohol

SCC: Monitoreo del consumo de calorías

FAF: Frecuencia de actividad física

TUE: Tiempo de uso de dispositivos tecnológicos

MTRANS: Tipo de transporte

NObesity: con los valores de: Peso Insuficiente, Peso Normal, Sobrepeso Nivel I, Sobrepeso Nivel II, Obesidad Tipo I, Obesidad Tipo II y Obesidad Tipo III

# Preparación de los datos

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X = df.drop('Tipo_obesidad', axis=1)
y = df['Tipo_obesidad']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42,stratify=y)
X_val,X_test,y_val,y_test= train_test_split(X_test, y_test, test_size=0.5, random_state=42,stratify=y_test)
```

```python
y_train.value_counts()
```

```python
scaler=StandardScaler() #z=(x-min)/sigma
X_train_Scaled=scaler.fit_transform(X_train)#calcula
X_val_Scaled=scaler.transform(X_val)
X_test_Scaled=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report, ConfusionMatrixDisplay
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Entrenamiento
logistic_model = LogisticRegression(max_iter=20)
logistic_model.fit(X_train_Scaled,y_train)
```

```python
# Predicciones
y_train_pred = logistic_model.predict(X_train_Scaled)
y_val_pred = logistic_model.predict(X_val_Scaled)
y_test_pred = logistic_model.predict(X_test_Scaled)
```

```python
# Métricas
print("Regresión Logistica")
print(f'Accuracy Entrenamiento: {accuracy_score(y_train, y_train_pred):.4f}')
print(f'Accuracy Validacion: {accuracy_score(y_val, y_val_pred):.4f}')
print(f'Accuracy Prueba: {accuracy_score(y_test, y_test_pred):.4f}\n')
print("Reporte de Clasificacion - Test")
print(classification_report(y_test, y_test_pred))
```

```python
# Matriz de confusión
cm_log = confusion_matrix(y_test, y_test_pred)
disp_log = ConfusionMatrixDisplay(confusion_matrix=cm_log)
disp_log.plot(cmap=plt.cm.Blues)
plt.title('Matriz de Confusión - Regresión Logística')
plt.show()
```

```python
#Arbol de decision
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
from sklearn.tree import plot_tree
```

```python
arbol = DecisionTreeClassifier(
    max_depth=10,
    min_samples_split=8,
    min_samples_leaf=2,
    random_state=34
)


arbol.fit(X_train, y_train)
```

```python
# Predicciones
y_train_pred_tree = arbol.predict(X_train)
y_val_pred_tree = arbol.predict(X_val)
y_test_pred_tree = arbol.predict(X_test)
```

```python
# Métricas
print("Arbol de Decision")
print(f'Accuracy Train: {accuracy_score(y_train, y_train_pred_tree):.4f}')
print(f'Accuracy Validation: {accuracy_score(y_val, y_val_pred_tree):.4f}')
print(f'Accuracy Test: {accuracy_score(y_test, y_test_pred_tree):.4f}\n')

print("Reporte de Clasificacion")
print(classification_report(y_test, y_test_pred_tree))
```

```python
# Matriz de Confusión
cm_test = confusion_matrix(y_test, y_test_pred_tree)
disp = ConfusionMatrixDisplay(confusion_matrix=cm_test)
disp.plot(cmap=plt.cm.Blues)
plt.title("Matriz de Confusión Árbol de Decisión")
plt.show()
```

```python
plt.figure(figsize=(80,20))
plot_tree(arbol, feature_names=X.columns, filled=True, rounded=True)
plt.title("Árbol de Decisión (max_depth=5)")
plt.show()
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlp_model = MLPClassifier(
    hidden_layer_sizes=(50,),
    activation='relu',
    batch_size=128,
    learning_rate_init=0.001,
    max_iter=500,
    solver='adam',
    alpha=0.01,
    early_stopping=True,
    validation_fraction=0.1,
    n_iter_no_change=10,
    random_state=42
)


mlp_model.fit(X_train_Scaled, y_train)
```

```python
# Predicción y evaluación
y_test_pred = mlp_model.predict(X_test_Scaled)
y_val_pred = mlp_model.predict(X_val_Scaled)
y_train_pred_MLP = mlp_model.predict(X_train_Scaled)
accuracy_mlp = accuracy_score(y_test, y_test_pred)
print(f'Accuracy entrenamiento - MLPClassifier: {accuracy_score(y_train, y_train_pred):.4f}')
print(f'Accuracy Test - MLPClassifier: {accuracy_mlp:.4f}')
print(f'Accuracy Validation - MLPClassifier: {accuracy_score(y_val, y_val_pred):.4f}')
print(classification_report(y_test, y_test_pred))
```

```python
# Matriz de confusión
cm_mlp = confusion_matrix(y_test, y_test_pred)
disp_mlp = ConfusionMatrixDisplay(confusion_matrix=cm_mlp)
disp_mlp.plot(cmap=plt.cm.Blues)
plt.title("Matriz de Confusión - MLPClassifier")
plt.show()
```

```python
#random forest
from sklearn.ensemble import RandomForestClassifier

clf_rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=8,
    min_samples_split=20,
    min_samples_leaf=19,
    max_features="sqrt",
    random_state=10
)
clf_rf.fit(X_train_Scaled, y_train)
```

```python
y_train_pred_clf = clf_rf.predict(X_train_Scaled)
y_val_pred_clf = clf_rf.predict(X_val_Scaled)
y_test_pred_clf = clf_rf.predict(X_test_Scaled)
```

```python
print("random forest")
print(f'Accuracy Train: {accuracy_score(y_train, y_train_pred_clf):.4f}')
print(f'Accuracy Validation: {accuracy_score(y_val, y_val_pred_clf):.4f}')
print(f'Accuracy Test: {accuracy_score(y_test, y_test_pred_clf):.4f}\n')

print("Reporte de Clasificacion")
print(classification_report(y_test, y_test_pred_clf))
```

```python
# Matriz de Confusión
cm_clf = confusion_matrix(y_test, y_test_pred_clf)
disp_clf = ConfusionMatrixDisplay(confusion_matrix=cm_clf)
disp_clf.plot(cmap=plt.cm.Blues)
plt.title("Matriz de Confusion - random forest")
plt.show()
```
